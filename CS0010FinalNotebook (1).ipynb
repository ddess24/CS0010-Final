## Parks by Neighborhood

The first dataset we analyzed contains entries for all the parks you can find in the city of Pittsburgh, including their name, type, and size, but we're only worried about the count per neighborhood.

As a result, we isolated the entries in the dataset containing parks with the type "Park", as there were many entries that included such as traffic dividers and other grassy areas too small to feasibly utilise for staying active.

Next, we counted how many times there was a park in each neighborhood, and decided to use a bar plot, because it's effective at fitting many datapoints in a small space and allows us to visually compare the datapoints with ease.

```python
# Import Pandas and MatPlotLib Modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Dataset of City of Pittsburgh Parks from WPRDC
parks = pd.read_csv("parks.csv")

# Isolate "type" and "neighborhood" Columns
parksShort = parks[["type","neighborhood"]]

# Remove Entries not Listed as "Parks"
mask = (parksShort["type"] == "Park")
parksFiltered = parksShort[mask]
parksFiltered.dropna()

# Sum Number of Parks by Neighborhood
numParksByNeighborhood = {}
for neighborhood in parksFiltered["neighborhood"]:
    if neighborhood in numParksByNeighborhood.keys():
        numParksByNeighborhood[neighborhood] += 1
    else:
        numParksByNeighborhood[neighborhood] = 1
        
# Sort Dictionary by Value
numParksByNeighborhood = dict(sorted(numParksByNeighborhood.items(), key=lambda x: x[1]))

# Create Bar Chart
fig = plt.subplots(figsize = (10,15))
neighborhoods = list(numParksByNeighborhood.keys())
numParks = list(numParksByNeighborhood.values())
plt.barh(neighborhoods, numParks, color="green")
plt.xlabel("Number of Parks")
plt.ylabel("Neighborhood")
plt.title("Number of Parks by Neighborhood")
plt.show()
```

## Parks Conclusion

Southside Slopes and Beechview happen to have the most parks in Pittsburgh, both boasting a total of 6.

Parks' versatility when it comes to staying active is a a huge benefit for soccer moms, so having plenty of them has Southside Slopes and Beechview in the lead. However, a certain contender seems to be in only fourth place...

# Baseball Fields by Neighborhood

For many young kids, baseball is a fun sport to play, so finding a neighborhood which can support the sport is important to our quest.

Our second dataset includes entries for these baseball fields and notes the neighborhoods they reside in, so we isolated the column containing these neighborhood names and counted how many times each neighborhood showed up.

We again used a bar graph to represent the data for its ease of visual comprehension.

```python
# Import Pandas and MatPlotLib Modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Dataset of City of Pittsburgh Baseball Fields from WPRDC
baseballFields = pd.read_csv("baseball.csv")

# Isolate Neighborhood Column
baseballFieldsShort = baseballFields.iloc[:,3].dropna()

# Sum Number of Baseball Fields by Neighborhood
numBaseballFieldsByNeighborhood = {}
for neighborhood in baseballFieldsShort:
    if neighborhood in numBaseballFieldsByNeighborhood.keys():
        numBaseballFieldsByNeighborhood[neighborhood] += 1
    else:
        numBaseballFieldsByNeighborhood[neighborhood] = 1

# Sort Dictionary by Value
numBaseballFieldsByNeighborhood = dict(sorted(numBaseballFieldsByNeighborhood.items(), key=lambda x: x[1]))

# Create Bar Chart
fig = plt.subplots(figsize = (10,15))
neighborhoods = list(numBaseballFieldsByNeighborhood.keys())
numBaseballFields = list(numBaseballFieldsByNeighborhood.values())
plt.barh(neighborhoods, numBaseballFields, color="red")
plt.xlabel("Number of Baseball Fields")
plt.ylabel("Neighborhood")
plt.title("Number of Baseball Fields by Neighborhood")
plt.show()
```

## Baseball Fields Conclusion

Brighton Heights and Squirrel Hill South are tied for first, having 7 baseball fields each, making them good choices for aspiring sports moms like ourselves, obviously.

Squirrel Hill South has placed rather high in the tlast two categories, so let's see if they can keep it up with our final category.

# Courts/Rinks by Neighborhood

Seemingly most important of all for soccer moms to look for in a neighborhood is the sports courts and rinks for sports like basketball, tennis, hockey, etc.

In this dataset, very similarly to the last, we isolated the neighborhood column for each entry, tallying up the occurrences of each neighborhood and graphing it with... you guessed it, a bar graph.

```python

# Import Pandas and MatPlotLib Modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Dataset of City of Pittsburgh Sports Courts and Rinks from WPRDC
courts = pd.read_csv("courts.csv")

# Isolate Neighborhood Column
courtsShort = courts.iloc[:,9].dropna()

# Sum Number of Courts by Neighborhood
numCourtsByNeighborhood = {}
for neighborhood in courtsShort:
    if neighborhood in numCourtsByNeighborhood.keys():
        numCourtsByNeighborhood[neighborhood] += 1
    else:
        numCourtsByNeighborhood[neighborhood] = 1

# Sort Dictionary by Value
numCourtsByNeighborhood = dict(sorted(numCourtsByNeighborhood.items(), key=lambda x: x[1]))

# Create Bar Chart
fig = plt.subplots(figsize = (10,15))
neighborhoods = list(numCourtsByNeighborhood.keys())
numCourts = list(numCourtsByNeighborhood.values())
plt.barh(neighborhoods, numCourts, color="gray")
plt.xlabel("Number of Courts/Rinks")
plt.ylabel("Neighborhood")
plt.title("Number of Courts/Rinks by Neighborhood")
plt.show()
```

## Courts Conclusion

And would you believe it, Squirrel Hill South knocks it out of the court in this category, with a total of 26 courts and rinks. Outstanding.
Notably, Highland Park came in second with 20 courts, but its no competition for Squirrel Hill South's domination.
