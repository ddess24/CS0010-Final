```python
# Import Pandas and MatPlotLib Modules
import pandas as pd
import matplotlib.pyplot as plt

# Import Dataset of City of Pittsburgh Parks from WPRDC
parks = pd.read_csv("parks.csv")

# Isolate "type" and "neighborhood" Columns
parksShort = parks[["type","neighborhood"]]

# Remove Entries not Listed as "Parks"
mask = (parksShort["type"] == "Park")
parksFiltered = parksShort[mask]
parksFiltered.dropna()

# Sum Number of Parks by Neighborhood
numParksByNeighborhood = {}
for neighborhood in parksFiltered["neighborhood"]:
    if neighborhood in numParksByNeighborhood.keys():
        numParksByNeighborhood[neighborhood] += 1
    else:
        numParksByNeighborhood[neighborhood] = 1
        
# Sort Dictionary by Value
numParksByNeighborhood = dict(sorted(numParksByNeighborhood.items(), key=lambda x: x[1]))

# Create Bar Chart
fig = plt.subplots(figsize = (10,15))
neighborhoods = list(numParksByNeighborhood.keys())
numParks = list(numParksByNeighborhood.values())
plt.barh(neighborhoods, numParks, color="green")
plt.xlabel("Number of Parks")
plt.ylabel("Neighborhood")
plt.title("Number of Parks by Neighborhood")
plt.show()
```
